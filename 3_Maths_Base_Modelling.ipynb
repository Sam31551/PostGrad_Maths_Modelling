##ZIP Model

```python
# Add a constant to the predictor features matrix (for the intercept)
X_train = sm.add_constant(X_train)
X_val = sm.add_constant(X_val)

# Initialize and fit the Zero-Inflated Poisson model
zip_model = sm.ZeroInflatedPoisson(endog=y_train, exog=X_train, exog_infl=X_train, inflation='logit')
zip_result = zip_model.fit()

# Predict the values for the validation set
predictions = zip_result.predict(exog=X_val, exog_infl=X_val)

# Calculate the evaluation metrics
mae = mean_absolute_error(y_val, predictions)
mse = mean_squared_error(y_val, predictions)
rmse = sqrt(mse)
r2 = r2_score(y_val, predictions)

print(f'MODEL: ZIP')
print('Mean Absolute Error: ', mae)
print('Mean Squared Error:', mse)
print('RMSE:', rmse)
print('R2:', r2)
```

```python
model_eval.append({'Model': zip_model, 'Mean Absolute Error': mae, 'Mean Squared Error': mse, 'RMSE': rmse,'R-squared': r2})
```

##Rest of models

The remaining models are all trained similarly. This means they can be looped through in order to streamline the model training process.

```python
models = [linear_regression_model, ridge_model, lasso_model, dt_model, rf_model, SVM, mlp, gam]
```

```python
from math import sqrt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import r2_score
```

Each model is trained and evaluated.

```python
for model in models:
  model.fit(X_train, y_train)
  predictions = model.predict(X_val)
  type(predictions)
  # Calculate the absolute errors
  errors = abs(predictions - y_val.values)
  # Calculate model evaluation metrics
  mae = mean_absolute_error(y_val, predictions)
  mse = mean_squared_error(y_val, predictions)
  rmse = sqrt(mse)
  r2 = r2_score(y_val, predictions)
  print(f'MODEL: {model}')
  print('Mean Absolute Error: ', mae, ' degrees.')
  print('Mean Squared Error:', mse, ' degrees')
  print('RMSE:', rmse, ' degrees')
  print('R2:', r2)
  model_eval.append({'Model': model, 'Mean Absolute Error': mae, 'Mean Squared Error': mse, 'RMSE': rmse,'R-squared': r2})
```
